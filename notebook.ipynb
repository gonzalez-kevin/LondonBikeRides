# Transformations

```python
# dict of more descriptive column names
new_col_names = {
    'timestamp': 'time', 
    'cnt': 'count', 
    't1': 'actual_temp_C',
    't2': 'feels_like_temp_C',
    'hum': 'humidity_pct',
    'wind_speed': 'wind_speed_kph',
    'weather_code': 'weather', 
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend', 
    'season': 'season'
}

bikes.rename(new_col_names, axis = 1, inplace = True)
```

```python
# transforming humidity into a percentage
bikes['humidity_pct'] = bikes['humidity_pct'] / 100
```

```python
# create mapping float values of season to written values
season_dict = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'autumn', 
    '3.0': 'winter'
}

# create mapping float values of weather to written values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}


# map weather and season columns
bikes['weather'] = bikes['weather'].astype('str').map(weather_dict)
bikes['season'] = bikes['season'].astype('str').map(season_dict)

bikes.head()
```

```python
# write bikes dataframe to an Excel file for further exploration
bikes.to_excel('london_bike_rides.xlsx', sheet_name = 'Date')
```
